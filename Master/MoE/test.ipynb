```python
import os
import json

# Path to the base folder with all datasets
base_path = "../all_datasets_global_by_years"

# Choose the excluded-state folder you want to inspect
excluded_state = "excluding_sp"        # example
state_folder = os.path.join(base_path, excluded_state)

if not os.path.exists(state_folder):
    raise FileNotFoundError(f"Folder {state_folder} does not exist")

print(f"Inspecting datasets in: {state_folder}\n")

# Iterate through all yearly dataset files
for dataset_file in sorted(os.listdir(state_folder)):
    if dataset_file.endswith(".jsonl"):
        dataset_path = os.path.join(state_folder, dataset_file)
        print(f"Dataset: {dataset_file}")

        # Count points in each sequence line
        with open(dataset_path, "r") as f:
            for idx, line in enumerate(f, start=1):
                data = json.loads(line)
                seq_len = len(data.get("sequence", []))
                print(f"  Line {idx:4d}: {seq_len} points")

        print()
```
